# Bar Graph

```python
# Plotting the bar chart from Plotly

fig = px.bar(
    df3, 
    x="Country Name", 
    y="value", 
    title="2020 Life Expectancy in Central African Republic, Germany, Jamaica, Japan", 
    color="Country Name",                                                       #"Country Name"  #low income country has low life expectancy  
    template="plotly",                                                          #templets available are: presentation,ggplot2, seaborn, simple_white, plotly, plotly_white, plotly_dark, xgridoff,ygridoff, gridon,none
    color_discrete_map={'CAF':'yellow','DEU': 'blue','JPN':'green','JAM':'pink'},
    text='value',                                                               #which is equal to the life expectancy value
    hover_data=['Region']
)

# fig.update_layout(legend={'x':0,'y':1.0}),                                    # defines where to keep the legend
fig.update_traces(texttemplate='%{text:.3s}', textposition='outside',width=[.5,.5,.5,.5])  #width of the bar is controlled



# fig.update_layout(showlegend=False)                                           # to remove the legend from the graph

fig.show()
```

### Using all countries, and picking a year 2020, make a scatter plot with x being GDP per Capita and y being Life Expentancy. This plot will tell you the relationship between wealth and health.

```python
my_indicators = ["SP.DYN.LE00.IN", "NY.GDP.PCAP.PP.CD"]
df_1 = df[df["indicator"].isin(my_indicators)]
df_1.shape
```

```python
df_2 = df_1.query('Year == 2020')
df_2
```

```python
df_3= df_2[["Year","value","indicator","Country Code","Country Name"]]
df_3.sample(5)
```

```python
df_4 = pd.pivot_table(df_3, values = 'value', index=['Country Code', 'Country Name','Year'], columns = 'indicator').reset_index()
df_4.sample(10)
```

```python
# Dropping the null values from the dataframe
df_4.dropna(inplace=True)
df_4.sample(10)
```

# Scatter Plot Wealth vs Health

```python
fig = px.scatter(
    df_4,
    x="NY.GDP.PCAP.PP.CD",
    y="SP.DYN.LE00.IN", 
    title="Wealth vs Health", 
    color="Country Name", 
    template="presentation"
)
fig.update_layout(showlegend=True)
fig.update_layout(
    xaxis_title="GDP per Capita",
    yaxis_title="Life Expectancy",)
fig.update_layout(showlegend=False)
fig.show()
```
