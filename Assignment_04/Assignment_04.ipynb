- Calculate the min, max, and average in-state tution of the above four institutions.

```python
# storing the above four universities in a dataframe named "stats"
stats= data[(data['INSTNM']=='Johns Hopkins University') | (data['INSTNM']=='Loyola University Maryland') 
            | (data['INSTNM']=='University of Maryland-College Park') 
            | (data['INSTNM']=='University of Maryland-Baltimore County') ]
```

```python
# displaying the stats dataframe
stats
```

```python
# Finding the minimum value in the column "TUTIONFEE_IN"
stats['TUITIONFEE_IN'].min()
```

```python
# Finding the maximum value in the column "TUTIONFEE_IN"
stats['TUITIONFEE_IN'].max()
```

```python
# Finding the average value in the column "TUTIONFEE_IN"
stats['TUITIONFEE_IN'].mean()
```
