```python
import pandas as pd
import plotly.express as px
```

```python
#Importing the data from the github by reading the csv file
df = pd.read_csv("https://raw.githubusercontent.com/srisaisrikar/Data_690_assignments/main/data690_world_dev/wdx_data.csv",index_col=0)
```

```python
df.shape
```

```python
df.head(2)
```

```python
# filtering GDP of BRICS countries between the period 2016-2021
gdp=df[(df['indicator']=='NY.GDP.PCAP.PP.CD')]
```

```python
gdp.sample(5)
```

```python
countries_gdp=gdp[(gdp['Country Name']=='China') | (gdp['Country Name']=='India') | (gdp['Country Name']=='Brazil') | (gdp['Country Name']=='South Africa') | (gdp['Country Name']=='Russian Federation')]
```

```python
gdp_line_graph=countries_gdp[['Year','value','Country Name']]
```

## GDP of BRICS during the period 2016-2020
- The line graph shows the growth of GDP for five nations through the years 2016-2020

```python
fig = px.line(gdp_line_graph[(gdp_line_graph['Year']==2016) | (gdp_line_graph['Year']==2017) | (gdp_line_graph['Year']==2018)|(gdp_line_graph['Year']==2019)| (gdp_line_graph['Year']==2020)], x="Year", y="value",color='Country Name', title='GDP per capita, PPP(current international $)',template ='plotly_dark')


fig.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 2016,
        dtick = 1
    )
)


fig.show()
```

## Is Foreign Direct investment creating employement ?
Lets look into this by taking into consideration the indicators as Foreign direct investment inflows(% of GDP), GDP per capita,PPP (current international $), and the Unemployement rate  

```python
# filtering the indicator foreign direct investment ('BX.KLT.DINV.WD.GD.ZS') from the data.
fdi=df[df['indicator']=='BX.KLT.DINV.WD.GD.ZS']
```

```python
fdi.head(2)
```

## Line chart of Foreign Direct Investment inflows(% of GDP), between the years 2016-2020  


```python
fig = px.line(fdi, x="Year", y="value",color='Country Name', title='FDI, net inflows(% of GDP)',template ='plotly_dark')

fig.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 2016,
        dtick = 1
    )
)
fig.show()  
```

## Bubble Chart between Unemployement, GDP and Foreign direct investment inflows. 

```python
# filtering the necessary parameters like FDI, GDP and Unemployment rates in the BRICS

df_graph=df[(df['indicator']=='BX.KLT.DINV.WD.GD.ZS') | (df['indicator']=='NY.GDP.PCAP.PP.CD') | (df['indicator']=='SL.UEM.TOTL.ZS') ]
```

```python
df_graph.head(2)
```

```python
# Pivoting the unique indicator values into columns for a better representation on the bubble chart

df_pivot_2016=pd.pivot_table(df_graph,values = 'value', 
                                                index=['Country Code', 'Country Name','Year'], 
                                                columns = 'indicator').reset_index()
df_pivot_2016.head(2)
```

```python
# since the dataset contains the null values, they are dropped.
df_pivot_2016.dropna(inplace=True)
```

```python
# Plotting the Bubble chart for year 2016 - 2020 

fig = px.scatter(
    df_pivot_2016,
    x="NY.GDP.PCAP.PP.CD",
    y="SL.UEM.TOTL.ZS", 
    animation_frame="Year",
    animation_group="Country Name",
    title="Bubble Chart Between Unemployment, FDI and GDP", 
    color="Country Name", 
    template="plotly_dark",
    size = "BX.KLT.DINV.WD.GD.ZS",
    # width=1000, 
    height=800,
    range_x=[0,40000],
    range_y=[0,50]

)

fig.update_layout(showlegend=True)
fig.update_layout(
    xaxis_title="GDP per capita, PPP (current international $)",
    yaxis_title="Unemployment total (% of total labor force) (modeled ILO estimate)",)

fig.show()
```

## Line chart of Unemployement 2016-2020

This represents the trend of unemployment of BRICS nation between the time period 2016-2020

```python
# filtering the data of unemployment indicator from the main dataframe

df_unemployment=df[df['indicator']=='SL.UEM.TOTL.ZS']
```

## Pie chart of Unemployment of South Africa vs The Rest

The below pie chart explains the ratio of unemployment in South Africa compared to the sum of unemployment rates in other BRICS countries (Brazil,Russia,India,China). 

```python
df_unemployment.head(2)
```

```python
df_unemployment=df_unemployment[df_unemployment['Year']==2020]
df_unemployment
```

```python
#summing up all the unemployment rates of other BRICS countries against South Africa
df_unemployment['rest']=df_unemployment['value'][:-1].sum()
df_unemployment.head(2)
```

```python
# the value of summation of other BRICS nations added to the dataframe in the form of a dict.
rest={'Country Name':'Rest','value':32.277}
df_unemployment=df_unemployment.append(rest,ignore_index=True)
```

```python
df_unemployment.drop(columns='rest',inplace=True)
pie_chart=df_unemployment[4:]
```

```python
# data required for the pie chart
pie_chart[['Country Name','value']]
```

```python
#Plotting the pie chart

fig = px.pie(
    pie_chart,
    names="Country Name", 
    values="value", 
    title="Unemployment South Africa vs The Rest of BRICS (2020)", 
    color="Country Name", 
    template="plotly_dark"
)

fig.update_layout(showlegend=True)

fig.update_traces(textposition='inside', textinfo='label+percent+value')

fig.show()
```

## Basic Drinking Water services among the BRICS nations

```python
# filtering the data required from the main dataframe
water_gdp=df[(df['indicator']=='SH.H2O.BASW.ZS') | (df['indicator']=='NY.GDP.PCAP.PP.CD') ]
```

```python
water_gdp=water_gdp[(water_gdp['Year']==2016) | (water_gdp['Year']==2017) | (water_gdp['Year']==2018) | (water_gdp['Year']==2019) | (water_gdp['Year']==2020)]
```

```python
water_gdp.head(2)
```

```python
df_pivot_water=pd.pivot_table(water_gdp,values = 'value', 
                                                index=['Country Code', 'Country Name','Year'], 
                                                columns = 'indicator').reset_index()
df_pivot_water.head(2)
```

```python
#Plotting the scatter plot for year 2016 - 2020 

fig = px.scatter(
    df_pivot_water,
    x="NY.GDP.PCAP.PP.CD",
    y="SH.H2O.BASW.ZS", 
    animation_frame="Year",
    animation_group="Country Name",
    title="Scatter Chart Between People using atleast basic drinking water services and GDP", 
    color="Country Name", 
    template="plotly_dark",
    # size = "BX.KLT.DINV.WD.GD.ZS",
    # width=1000, 
    height=800,
    range_x=[0,35000],
    range_y=[85,100]

)

fig.update_layout(showlegend=True)
fig.update_layout(
    xaxis_title="GDP per capita, PPP (current international $)",
    yaxis_title="People using atleast basic drinking water services (% of population)",)

fig.show()
```

## Agricultural Land vs GDP vs Cereal Production

```python
agri_gdp_ceral.head(2)
```

```python
#Plotting the Bubble chart for years 2016 - 2020 

fig = px.scatter(
    df_pivot_agri[(df_pivot_agri['Year']==2016) | (df_pivot_agri['Year']==2017) |(df_pivot_agri['Year']==2018) |(df_pivot_agri['Year']==2019)|(df_pivot_agri['Year']==2020)],
    x="AG.LND.AGRI.ZS",
    y="AG.PRD.CREL.MT", 
    animation_frame="Year",
    animation_group="Country Name",
    title="Agri land vs Ceral Production vs GDP", 
    color="Country Name", 
    template="plotly_dark",
    size = "NY.GDP.PCAP.PP.CD",
    # width=1000, 
    height=800,
    # range_x=[0,40000],
    # range_y=[0,50]

)

fig.update_layout(showlegend=True)
fig.update_layout(
    xaxis_title="Agricultural Land (% of land area)",
    yaxis_title="Ceral Production (metric ton)",)

fig.show()
```

## Conclusion

A cross country comparison for the BRICS shows that economic progress alone may not necessarily translate into a higher quality of life for these economies. The countries which lag behind some in GDP per capita, are more progressive in some indicators. Since progress in every sector of a nation may also affect its economic prosperity, it is crucial for nations to undertake measures in the all major spheres in which they are lagging behind. While India still needs to invest its resources in meeting its basic human needs, countries such as China and Russia need to attract foreign investments which can boost their economy as well as provide employment
